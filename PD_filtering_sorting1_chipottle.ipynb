# Filtering and Sorting Data

```python
import pandas as pd
```

### Dataset in this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv)

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')

print(chipo.shape)
chipo.head()
```

### Number of products > $10.00

```python
# Strip symbols in-place and convert to float
chipo['item_price'] = chipo['item_price'].str.strip('$').astype(float)

# Test stripping and data type
print(chipo.iloc[0][4])
print(chipo['item_price'].dtype)
```

```python
chipo[chipo['item_price'] > 10].count()
```

### Price of each item

```python
items_prices = chipo[['item_name', 'item_price']]
items_prices.head()
```

### Sort by itemÂ´s name (alphabetically)

```python
items_prices.sort_values(by='item_name')
```

### Quantity of the most expensive item ordered

```python
chipo.sort_values(by='item_price', ascending=False).iloc[0]['quantity']

# OR

chipo.sort_values(by='item_price', ascending=False).head(1)
```

### Number of Veggie Salad Bowls ordered

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl'].iloc[:]['quantity'].sum()
```

### Number of times that people orderd more than one Canned Soda

```python
chipo[(chipo['item_name'] == "Canned Soda") & (chipo['quantity'] > 1)].shape[0]
```
